```python
from pymongo import MongoClient
import pandas as pd

client = MongoClient("mongodb://localhost:27017/")

db = client.db_companies # crear una base de datos

collection_companies = db.companies

```

# Answers
### 1. All the companies that it's name match 'Babelgum'. Retrieve only their `name` field.

```python
list(db.companies.find({'name':'Babelgum'},{'_id': 0 , 'name':1}))
```

### 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by **number of employees**.

- **`query`**: /*find{number_of_employees: {'$gt': 5000}}*/
- **`sort`**: /*{number_of_employees : 1}*/
- **`limit`**: /*20*/

```python
list(db.companies.find({'number_of_employees': {'$gt':5000}}).limit(20).sort('number_of_employees', 1))[0]
```

### 3. All the companies founded between 2000 and 2005, both years included. Retrieve only the `name` and `founded_year` fileds.

- **`query`**: /*You should copy/paste the query in here*/
- **`projection`**: /*You should copy/paste the projection in here*/
- **`sort`**: /*You should copy/paste the sort in here*/
- **`skip`**: /*You should copy/paste the skip in here*/
- **`limit`**: /*You should copy/paste the limit in here*/

```python
list(db.companies.find(
  { "$and": [{"founded_year": {"$gte": 2000}}, {"founded_year":{"$lte": 2005}}] }, 
  { "name": 1, "founded_year": 1, "_id": 0 }
))[:20]
```

### 4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the `name` and `ipo` fields.

```python
list(db.companies.find(
  { '$and': [{"ipo.valuation_amount": {'$gt': 10000000}}, {'founded_year':{'$lt': 2010}}] }, 
  {'name': 1, 'ipo': 1, '_id': 0}
))[:20]
```

### 5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.

```python
list(db.companies.find({'$and':[{'number_of_employees':{'$lt':1000}},{'founded_year':{'$lt':2005}}]}).limit(10).sort('number_of_employees', 1))[1] 
```

### 7. All the companies that have a null type of value on the `category_code` field.

### 8. All the companies that have at least 100 employees but less than 1000. Retrieve only the `name` and `number of employees` fields.

```python
list(db.companies.find(
  { '$and': [{'number_of_employees': {'$gte': 100}}, {'number_of_employees': {'$lt': 1000}}]}, 
  {'name': 1, 'number_of_employees': 1, '_id': 0}
))[0]
```
